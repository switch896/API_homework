```python
import pandas as pd
import openweathermapy.core as owm
import json
import random
import numpy as np



with open("city_id.json", encoding = "utf8") as data_file:
    city_list = json.load(data_file)
bin1 = []
bin2 = []
bin3 = []    
bin4 = []    
bin5 = []    
bin6 = []    
bin7 = []    
bin8 = []
bin9 = []

 


for city_id in city_list:
    if abs(city_id["coord"]["lat"])<10:
        bin1.append(city_id["id"])
    elif abs(city_id["coord"]["lat"]) >= 10 and abs(city_id["coord"]["lat"]) <20:
        bin2.append(city_id["id"])
    elif abs(city_id["coord"]["lat"]) >= 20 and abs(city_id["coord"]["lat"]) <30:
        bin3.append(city_id["id"])
    elif abs(city_id["coord"]["lat"]) >= 30 and abs(city_id["coord"]["lat"]) <40:
        bin4.append(city_id["id"])
    elif abs(city_id["coord"]["lat"]) >= 40 and abs(city_id["coord"]["lat"]) <50:
        bin5.append(city_id["id"])
    elif abs(city_id["coord"]["lat"]) >= 50 and abs(city_id["coord"]["lat"]) <60:
        bin6.append(city_id["id"])
    elif abs(city_id["coord"]["lat"]) >= 60 and abs(city_id["coord"]["lat"]) <70:
        bin7.append(city_id["id"])
    elif abs(city_id["coord"]["lat"]) >= 70 and abs(city_id["coord"]["lat"]) <80:
        bin8.append(city_id["id"])
    elif abs(city_id["coord"]["lat"]) >= 80 and abs(city_id["coord"]["lat"]) <90:
        bin9.append(city_id["id"])
num_cities = len(city_list) 
sample_ids = []
sample_ids += random.sample(bin1,90)
sample_ids += random.sample(bin2,90)
sample_ids += random.sample(bin3,90)
sample_ids += random.sample(bin4,90)
sample_ids += random.sample(bin5,90)
sample_ids += random.sample(bin6,90)
sample_ids += random.sample(bin7,90)
sample_ids += bin8

key = "d0ae2cd96fb035aead5af43e60a12837"
base_url = "api.openweathermap.org/data/2.5/weather?"
settings = {"units": "imperial", "APPID": key}

city_id = []
city_name = []
temperature = []
humidity = []
cloudiness = []
latitude = []
wind_speed = []

for city in sample_ids:
    response = owm.get_current(city,**settings)
    humidity.append(response["main"]["humidity"])
    city_id.append(response["id"])
    city_name.append(response["name"])
    temperature.append(response["main"]["temp"])
    cloudiness.append(response["clouds"]["all"])
    latitude.append(response["coord"]["lat"])
    wind_speed.append(response["wind"]["speed"])
    print("City: "+ response["name"])
    print("City ID: " +str(response["id"]))
    url= base_url + "&id=" + str(city) +"&APPID=" + key + "&units=imperial"
    print(url)


```

```python
city_id = np.array(city_id)
city_name = np.array(city_name)
temperature = np.array(temperature)
humidity = np.array(humidity)
cloudiness = np.array(cloudiness)
latitude = np.array(latitude)
graph_data = pd.DataFrame(columns = ["city_id", "city_name", "temperature", "humidity", "cloudiness", "latitude","wind_speed"])
graph_data["city_id"] = city_id 
graph_data["city_name"] = city_name
graph_data["temperature"] = temperature
graph_data["humidity"] = humidity
graph_data["cloudiness"] = cloudiness
graph_data["latitude"] = latitude
graph_data["wind_speed"] = wind_speed
```
